```python
# import Numpy & Pandas
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/saarland 2021-01-05 to 2023-09-30.csv')
```

```python
df
```

```python
df.columns
```

```python
df1 = df.drop(columns=['name', 'tempmax', 'tempmin', 'feelslikemax', 'feelslikemin', 'feelslike', 'precipprob', 'precipcover', 'preciptype', 'snowdepth', 'windgust', 'severerisk', 'sunrise','sunset', 'moonphase', 'conditions', 'description', 'icon', 'stations' ])
```

```python
df2 = df1.isna()
df2
```

```python
df2 = df1.dropna()
```

```python
df2
```

```python
df2.isna().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = df2.temp, y = df2.dew, color = "r" )
```

```python
x = df2.temp
x = x.values.reshape(-1,1)
y = df2.dew
```

```python
x.shape
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.33, random_state= 10)
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg = reg.fit(x_train, y_train)

y_perd = reg.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
from math import sqrt
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_perd))
```

```python
print(rmse)
```

```python
mae = mean_absolute_error(y_test, y_perd)
print(mae)
```
